```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
```

```python
df_train=pd.read_csv('train.csv')
train=df_train.drop(columns=['RowNumber', 'CustomerId','Surname','Geography','Gender'])
df_test=pd.read_csv('test.csv')
test=df_test.drop(columns=['RowNumber', 'CustomerId','Surname','Geography','Gender'])

features=train.columns[:8]
y=train['Exited']

clf=RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[features],y)
clf.predict(test[features])

upload_df=pd.DataFrame()
upload_df['Unnamed: 0']=pd.Series(range(0,2000))
upload_df['RowNumber']=df_test['RowNumber']
upload_df['Exited']=clf.predict(test[features])
upload_df.to_csv('churn_prediction_data')
```
