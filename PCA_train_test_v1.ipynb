```python
import pandas as pd
data=pd.read_csv('train.csv')
train=data.drop(['RowNumber','CustomerId','Surname','Exited'],axis=1)
test=pd.read_csv('test.csv')
test=test.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df=pd.concat([train,test],axis=0)
df
```

## 特徵工程

## df

```python
df['BalanceSalaryRatio'] = df.Balance/df.EstimatedSalary
df['TenureByAge'] = df.Tenure/(df.Age)
df['CreditScoreGivenAge'] = df.CreditScore/(df.Age)
df.head()
```

### train data

```python
train['BalanceSalaryRatio'] = train.Balance/train.EstimatedSalary
train['TenureByAge'] = train.Tenure/(train.Age)
train['CreditScoreGivenAge'] = train.CreditScore/(train.Age)

train.loc[train.HasCrCard == 0, 'HasCrCard'] = -1
train.loc[train.IsActiveMember == 0, 'IsActiveMember'] = -1

# One hot encode the categorical variables
import numpy as np
lst = ['Geography', 'Gender']
remove = list()
for i in lst:
    if (train[i].dtype == np.str or train[i].dtype == np.object):
        for j in train[i].unique():
            train[i+'_'+j] = np.where(train[i] == j,1,-1)
        remove.append(i)
train = train.drop(remove, axis=1)
train.head()
```

```python
# minMax scaling the continuous variables
continuous_vars = ['CreditScore',  'Age', 'Tenure', 'Balance','NumOfProducts', 'EstimatedSalary','BalanceSalaryRatio',
                   'TenureByAge','CreditScoreGivenAge']
minVec = df[continuous_vars].min().copy()
maxVec = df[continuous_vars].max().copy()
train[continuous_vars] = (train[continuous_vars]-minVec)/(maxVec-minVec)
train.head()
```

### test data

```python
test['BalanceSalaryRatio'] = test.Balance/test.EstimatedSalary
test['TenureByAge'] = test.Tenure/(test.Age)
test['CreditScoreGivenAge'] = test.CreditScore/(test.Age)

test.loc[test.HasCrCard == 0, 'HasCrCard'] = -1
test.loc[test.IsActiveMember == 0, 'IsActiveMember'] = -1

# One hot encode the categorical variables
import numpy as np
lst = ['Geography', 'Gender']
remove = list()
for i in lst:
    if (test[i].dtype == np.str or test[i].dtype == np.object):
        for j in test[i].unique():
            test[i+'_'+j] = np.where(test[i] == j,1,-1)
        remove.append(i)
test = test.drop(remove, axis=1)
test.head()
```

```python
# minMax scaling the continuous variables
test[continuous_vars] = (test[continuous_vars]-minVec)/(maxVec-minVec)
test.head()
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 8)

X_train_pca=pca.fit_transform(train)
X_test_pca=pca.transform(test)
Y_train=data.Exited

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Fit Random Forest classifier
RF = RandomForestClassifier()
RF.fit(X_train_pca,Y_train)

print(classification_report(Y_train,  RF.predict(X_train_pca)))
#print(classification_report(Y_test,  RF.predict(X_test_pca)))
```
